6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
df_orders_products_merged_update['average_price'] = df_orders_products_merged_update.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_orders_products_merged_update.head(20)
```

```python
df_orders_products_merged_update.loc[df_orders_products_merged_update['average_price'] < 10, 'spending_flag'] = 'Low Spender'
df_orders_products_merged_update.loc[df_orders_products_merged_update['average_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
df_orders_products_merged_update['spending_flag'].value_counts(dropna = False)
```

```python
df_orders_products_merged_update[['user_id', 'spending_flag', 'loyalty_flag', 'order_number']].head(20)
```

7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column. The criteria for the flag should be as follows:If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.

```python
df_orders_products_merged_update['median_days_since_prior_order'] = df_orders_products_merged_update.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_orders_products_merged_update.head(20)
```

```python
df_orders_products_merged_update.loc[df_orders_products_merged_update['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_orders_products_merged_update.loc[(df_orders_products_merged_update['median_days_since_prior_order'] <= 20) & (df_orders_products_merged_update['median_days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
df_orders_products_merged_update.loc[df_orders_products_merged_update['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_orders_products_merged_update['order_frequency_flag'].value_counts(dropna = False)
```
